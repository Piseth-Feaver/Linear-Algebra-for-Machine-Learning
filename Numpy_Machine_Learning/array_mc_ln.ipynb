**How to use array Function?**

```python
from numpy import array
l = [1, 2, 3]
a = array(l)
print(a)
print(a.shape)
```

**How to use empty Function?**

```python
from numpy import empty
a = empty([3, 1])
print(a)
```

**How to use Zero Function?**

```python
from numpy import zeros
a = zeros([3,5])
print(a)
```

**How to use Ones Function?**

```python
from numpy import ones
a = ones([5,5])
print(a)
```

**How to remove float number to integer number ?**

```python
from numpy import array
a = ones([5])
b = a.astype(int)
print(b)
```

**How to use vstack Function?**

```python
from numpy import array
from numpy import vstack
a1 = array([1, 2, 3])
print(a1)
a2 = array([4, 5, 6])
# vstack function to combine array together like vertical line
a3 = vstack((a1, a2))
print(a3)
print(a3.shape)
```

**How to use hstack Fucntion? **
hstack combine array together like horizontal

```python
from numpy import array
from numpy import hstack
a1 = array([1, 2, 3])
a2 = array([4, 5, 6])
a3 = hstack((a1, a2))
print(a3)
print(a3.shape)
```

Two-Dimensional Slicing

```python

from numpy import array
data = array([
    [11, 22, 33],
    [44, 55, 66],
    [77, 88, 99],
])
x = data[ :, :-1]
y = data[:, -1]
print(x)
print(y)
```

```python
from numpy import array
data = array([
    [11, 22, 33],
    [44, 55, 66],
    [77, 88, 99],
])
split = 3
train, test= data[:split,:], data[split:,:]
print(train,test)
```

```python
from numpy import array

# 3D array
data_3d = array([
    [[11, 22, 33], [44, 55, 66]],  # First 2D array
    [[77, 88, 99], [111, 122, 133]],
    [[101, 102, 103], [104, 105, 106]]
])
print(data_3d)
print(data_3d.shape)
# How to read 3D in numpy 3 layers 1 layers has 2 row and 2 row has 3 column
```

```python
from numpy import array
data = array([11, 22, 33, 44, 55])
data = data.reshape((data.shape[0], 1))
print(data)
print(data.shape)
```
