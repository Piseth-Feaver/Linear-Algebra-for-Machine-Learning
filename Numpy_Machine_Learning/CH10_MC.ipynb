```python
from numpy import array
from numpy import triu
from numpy import tril
M = array([
    [1, 2, 3],
    [1, 2, 3],
    [1, 2, 3],
])
print(M)
lower = tril(M)
print(lower)
upper = triu(M)
print(upper)
```

```python
from numpy import array
from numpy import diag
M  = ([
    [1, 0, 9],
    [1, 9, 3],
    [2, 1, 0]
])
print(M)
d = diag(M)
print(d)
```

```python
# diag function
from numpy import array
from numpy import diag
M = ([
    [1, 2, 3],
    [1, 2, 3],
    [1, 2, 3],
])
print(M)
d = diag(M)
print(d)
D = diag(d)
print(D)
```

```python
# identity matrix
from numpy import identity
I = identity(4)
print(I)
```

```python
from numpy import array
from numpy.linalg import inv
Q = array([
    [1, 0],
    [0, -1],
])
print(Q)
V = inv(Q)
print(V)
```

```python
import numpy as np
m = np.array([
    [1, 0],
    [0, 3],
])
result = np.linalg.inv(m)
print(result)
```
