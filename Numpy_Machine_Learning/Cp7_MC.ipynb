**How to define a Vector?**

```python
from numpy import array
v = array([1, 2, 3])
print(v)
```


How to added Vector together?


```python
from numpy import array
a = array([1, 2, 3])
print(a)
b = array([4, 5, 6])
print(b)
c = a + b
print(c)
```

**How to use Vector Subtraction?**

```python
from numpy import array
a = array([1, 2, 3])
print(a)
b = array([0.5, 0.3, 4])
b = b.astype(int)
c = a - b
print(c)
```

**How to use Vector Multiplication?**

```python
from numpy import array
a = array([1, 2, 3])
print(a)
b = array([1, 2, 3])
print(b)
c = a * b
print(c)
```

**How to use Vector Division?**

```python
from numpy import array
a = array([1, 2, 3])
print(a)
b = array([1, 2, 3])
print(b)
c = a / b
print(c)
```

How to use Dot?

```python
from numpy import array
a = array([1, 2, 3])
print(a)
b = array([4, 5, 6])
print(b)
c = a.dot(b)
print(c)
```

```python
import numpy as np
a = ([1, 2, 3])
print(a)
b = ([4, 5, 6])
print(b)
c = np.dot(a, b)
print(c)
```

**How to use Vector scalar multiplication?**

```python
from numpy import array
a = array([1, 2, 3])
print(a)
b = 0.5
print(b)
c = a * b
print(c)
```

How to use Vector norm?

```python
from numpy import array
from numpy.linalg import norm
a = array([1, 2, 3])
print(a)
l1 = norm(a,1)
print(l1)
```
